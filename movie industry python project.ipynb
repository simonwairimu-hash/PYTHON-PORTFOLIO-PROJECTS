```python
# import libraries
import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize']=(12,8) #adjusting the configurations of the plot

#read in the data
df = pd.read_csv(r'C:\Users\user\Desktop\DATA ANALYSIS\FINISHED PROJECTS\PYTHON\movies.csv')
```

```python
# data 

df.head()
```

```python
# checking for missing data
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col,pct_missing))
```

```python
# data types
df.dtypes
```

```python
#change data type of columns

df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0).astype(int)

df['gross'] = pd.to_numeric(df['gross'], errors='coerce').fillna(0).astype(int) 
```

```python
df.head()
```

```python
#maatching year and release year

df['yearcorrect'] = df['released'].astype(str).str.split().str[2]

df.head()
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# dropping dublicates

df['company'].drop_duplicates().sort_values(ascending = False)
```

```python
df.head()
```

```python
# scatter plot with budget vs gross

plt.scatter(x=df['budget'], y=df['gross'])

plt.title('Budget vs Gross Earning')

plt.xlabel('Gross Earnings')

plt.ylabel('Budget for Film')

plt.show()
```

```python
df.head()
```

```python
# plot budget vs gross using searborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color": "red"}, line_kws={"color": "blue"})

plt.show()
```

```python
# checking correlation

df.corr(numeric_only=True, method='pearson')
```

```python
correlation_matrix = df.corr(numeric_only=True, method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```

```python
#company

df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(numeric_only=True, method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')

plt.xlabel('Movie Features')

plt.ylabel('Movie Features')

plt.show()
```
